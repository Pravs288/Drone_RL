```python
def play_one_episode(agent, eps, ue):
    obs = agent.env.reset(ue)
    iters = 0
    total_rwd = 0
    while (iters < 50):
        action = agent.sample_action(obs, eps)

        prev_obs = obs[:]

        obs, rwd = agent.env.step(action)

        agent.Update_Q(prev_obs, action, obs, rwd)

        total_rwd += rwd
        iters+=1

    return total_rwd
```

```python
if __name__ == "__main__":
    
    #omega_vec = [30, 45, 60, 90]
    #Agent Parameters
    env = RFBeamEnv()
    alpha = 0.1
    gamma =  0.99
    eps = 0.2
    
    #UE parameters
    ue_s = [10,15,0]
    ue_v = 10
    ue_tdist = 90
    possible_ue_loc = [[x,ue_s[1],ue_s[2]] for x in range(ue_s[0],ue_s[0]+ue_tdist,ue_v)]
    
    agent = RL_Agent(env, alpha, gamma)
    print("Before episode: {0}".format(agent.Q))
    M = 30 #no. of episodes
    ep_ndx = 0
    ep_rwds = []
    while (ep_ndx < M):
            rnd_ndx = np.random.randint(0, len(possible_ue_loc))
            ue_r = possible_ue_loc[rnd_ndx][:]
            
            play_one_episode(agent, eps, ue_r)
            ep_ndx += 1
    
    
    print("Training Done!!")
    #print("After episode: {0}".format(agent.Q))
```
